## PROBLEM no.2 

```python
#Import the data needed for coding
Board = pd.read_csv('board2.csv')

#Display the data for reference
Board
```

```python
#Create new coloumn for the average of all subjects to suffice the needed materials 
Board['Average'] = Board[['Math', 'Electronics', 'GEAS', 'Communication']].mean(axis=1)
Board
```

```python
#Calculate the average grades for each data set
AVE_Track = Board.groupby('Track')['Average'].mean()
AVE_Gender = Board.groupby('Gender')['Average'].mean()
AVE_Hometown = Board.groupby('Hometown')['Average'].mean()
```

```python
#Display the following by data set (First - Track)
AVE_Track.plot(kind='bar', color='skyblue', title='Average Grade by Track', xlabel='Track', ylabel='Average Grade')
plt.show()
```

```python
#Display the following by data set (Second - Track)
AVE_Gender.plot(kind='bar', color='lightgreen', title='Average Grade by Gender', xlabel='Gender', ylabel='Average Grade')
plt.show()
```

```python
#Display the following by data set (Last - Track)
AVE_Hometown.plot(kind='bar', 
                  color='lightcoral', 
                  title='Average Grade by Hometown', 
                  xlabel='Hometown', 
                  ylabel='Average Grade')
plt.show()
```
